```python
import cv2
```

```python
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/cat_image.jpg")

cv2_imshow(image)

```

```python
print("Shape:", image.shape)
# 350 => height
# 625 => width
# 3 => color channels (BGR) =>Blue, Green, Red
print("Height:",image.shape[0])
print("Width:",image.shape[1])
print("Channels:",image.shape[2])
```

```python
(b, g, r) = image[100, 50]
print("Blue:", b, "Green:", g, "Red:", r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(gray,(250, 250))
cv2_imshow(resized)
```

```python
cropped = gray[100:400, 100:400]
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(gray, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(gray, 1)
cv2_imshow(flipped)
```

```python
cv2.imwrite("cropped_cat.jpg", cropped)
```
